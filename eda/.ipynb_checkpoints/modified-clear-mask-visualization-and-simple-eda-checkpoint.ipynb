### Training data: train_images/ and train.csv

```python
train_df = pd.read_csv(Path(input_dir,'train.csv'),index_col='ImageId')
print(f"There are {len(train_df)} total labeled images in 'train.csv'.")
train_df.head()
```

```python
train_images = os.listdir(Path(input_dir,'train_images'))
print(f"There are {len(train_images)} total images in 'train_images/' directory.")
```

```python
# calculate number of duplicates of each labeled image
ImageId_duplicates = train_df.index.value_counts().value_counts()
n_imgs = 0
for n_dup in ImageId_duplicates.index:
    print(f'{ImageId_duplicates[n_dup]} images appear {n_dup} time(s).')
    n_imgs += ImageId_duplicates[n_dup]*n_dup
print(f'There are {n_imgs} total labeled images.')
```

```python
# frequency of classes
n_class = defaultdict(lambda: 0) # number of defects per class
n_defects = defaultdict(lambda: 0) # number of defects per images
n_combos = defaultdict(lambda: 0) # number of defects combinations
images_by_class = defaultdict(lambda: []) # store images by class type
for ImageId in train_images:
    try:
        class_types = train_df.loc[ImageId]['ClassId']
        if type(class_types) == np.int64:
            n_class[class_types] += 1
            n_defects[1] += 1
            images_by_class[str(class_types)].append(ImageId)
        else:
            for class_type in class_types:
                n_class[class_type] += 1
            n_defects[len(class_types)] += 1
            n_combos[tuple(list(class_types))] += 1
            images_by_class[str(tuple(list(class_types)))].append(ImageId)
    except:
        n_class[0] += 1
        n_defects[0] += 1
        images_by_class['None'].append(ImageId)
```

```python
# plot frequency of defect classes
fig, ax = plt.subplots(figsize=(10,5))
sns.barplot(x=['0','1','2','3','4'], y=list(n_class.values()), ax=ax)
ax.set_title("Frequency of Defect Classes")
labels = [item.get_text() for item in ax.get_xticklabels()]
labels[0] = 'None'
ax.set_xticklabels(labels)
ax.set_xlabel("Class Type")
ax.set_ylabel("Frequency")
ax.bar_label(ax.containers[0])
ax.set_ylim(0,6500)
plt.show()
```

```python
# frequency of number defect per image
fig, ax = plt.subplots(figsize=(10,5))
sns.barplot(x=['0','1','2','3'], y=list(n_defects.values()), ax=ax)
ax.set_title("Frequency of Number of Detects in each Image")
labels = [item.get_text() for item in ax.get_xticklabels()]
labels[0] = 'None'
ax.set_xticklabels(labels)
ax.set_xlabel("Number of Defects")
ax.set_ylabel("Frequency")
ax.set_ylim(0,6750)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# frequency of number defect per image
fig, ax = plt.subplots(figsize=(10,5))
keys = [str(key) for key in n_combos.keys()]
sns.barplot(x=keys, y=list(n_combos.values()), ax=ax)
ax.set_title("Frequency of Defect Combinations")
labels = [item.get_text() for item in ax.get_xticklabels()]
ax.set_xticklabels(labels)
ax.set_xlabel("Defect Combination")
ax.set_ylabel("Frequency")
ax.set_ylim(0,350)
ax.bar_label(ax.containers[0])
plt.show()
```

# Let's visualization masks!

```python
palet = [(249, 192, 12), (0, 185, 241), (114, 0, 218), (249,50,12)]
```

```python
def name_and_mask(start_idx):
    col = start_idx
    img_names = [str(i).split("_")[0] for i in train_df.iloc[col:col+4, 0].values]
    if not (img_names[0] == img_names[1] == img_names[2] == img_names[3]):
        raise ValueError

    labels = train_df.iloc[col:col+4, 1]
    mask = np.zeros((256, 1600, 4), dtype=np.uint8)

    for idx, label in enumerate(labels.values):
        if label is not np.nan:
            mask_label = np.zeros(1600*256, dtype=np.uint8)
            label = label.split(" ")
            positions = map(int, label[0::2])
            length = map(int, label[1::2])
            for pos, le in zip(positions, length):
                mask_label[pos-1:pos+le-1] = 1
            mask[:, :, idx] = mask_label.reshape(256, 1600, order='F')
    return img_names[0], mask
```

```python
idx_no_defect = []
idx_class_1 = []
idx_class_2 = []
idx_class_3 = []
idx_class_4 = []
idx_class_multi = []
idx_class_triple = []

for col in range(0, len(train_df), 4):
    img_names = [str(i).split("_")[0] for i in train_df.iloc[col:col+4, 0].values]
    if not (img_names[0] == img_names[1] == img_names[2] == img_names[3]):
        raise ValueError
        
    labels = train_df.iloc[col:col+4, 1]
    if labels.isna().all():
        idx_no_defect.append(col)
    elif (labels.isna() == [False, True, True, True]).all():
        idx_class_1.append(col)
    elif (labels.isna() == [True, False, True, True]).all():
        idx_class_2.append(col)
    elif (labels.isna() == [True, True, False, True]).all():
        idx_class_3.append(col)
    elif (labels.isna() == [True, True, True, False]).all():
        idx_class_4.append(col)
    elif labels.isna().sum() == 1:
        idx_class_triple.append(col)
    else:
        idx_class_multi.append(col)
```
